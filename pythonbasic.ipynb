### contact application
- add contact details
- search the contact details
- list all the contacts(merge/import)
- name 1 : value
- name 2 : value
- modify or update the contact details
- delete the contact details

```
    contacts={} # empty dictionary
    def addcontact(name,phone):
        if name not in contacts:
            contacts[name]=phone
            print('contact detais are added successfully')
        else:
            print('name already exists')
        return
    addcontact('kishore','93222142')
    addcontact('sai','9213452')
    addcontact('saketh','764992')
    addcontact('hemanth','9854788')
    
```

```
def modifycontact(name,phone):
    if name in contacts:
        
        contacts[name]=phone
        print(name," ",contacts[name])
        print('contact modified successfully')
    else:
        print("contact does not exist")
    return
modifycontact('hemanth')
```

```
def delcontact(name):
    if name in contacts:
        del contacts[name]
        print(name, 'deleted successfully')
    else:
        print(name, 'not present')
    return
delcontact('saketh')
```
